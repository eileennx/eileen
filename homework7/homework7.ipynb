```python
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 5))

for amp, freq in zip(amplitude, frequency):
    y1 = make_sine_wave(x, amp, 1)
    axs[0].plot(x, y1, label=f'Amp={amp}')
    axs[0].set_title('Varying Amplitude (Freq=1)', fontsize=16)
    
    y2 = make_sine_wave(x, 1, freq)
    axs[1].plot(x, y2, label=f'Freq={freq}')
    axs[1].set_title('Varying Frequency (Amp=1)', fontsize=16)
    
    y3 = make_sine_wave(x, amp, freq)
    axs[2].plot(x, y3, label=f'A={amp}, f={freq}')
    axs[2].set_title('Varying Both Amp and Freq', fontsize=16)

for ax in axs:
    ax.set_xlabel('x', fontsize=14)
    ax.set_ylabel('y', fontsize=14)
    ax.legend(fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
