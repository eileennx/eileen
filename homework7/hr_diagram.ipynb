```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
print(f"Data shape: {data.shape}")
print(f"There are {data.shape[0]} rows (stars) in this dataset.")
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
min_mag = np.min(magnitude) 
max_mag = np.max(magnitude)
print(f"Minimum Magnitude: {min_mag}")
print(f"Maximum Magnitude: {max_mag}")
```

```python
min_temp = np.min(temperature)
max_temp = np.max(temperature)
print(f"Minimum Temperature: {min_temp} K")
print(f"Maximum Temperature: {max_temp} K")
```

```python
brightest_star_mag = min_mag
dimmest_star_mag = max_mag
print(f"The brightest star has a magnitude of {brightest_star_mag}.")
print(f"The dimmest star has a magnitude of {dimmest_star_mag}.")
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, alpha=0.5, label='Stars', s=10)

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.scatter(np.log(min_temp), np.log(min_mag), color="orange", marker="*", s=300, label="Coldest Star")
plt.scatter(np.log(max_temp), np.log(max_mag), color="green", marker="*", s=300, label="Hottest Star")
plt.scatter(np.log(max_temp), np.log(min_mag), color="black", marker="*", s=300, label="Brightest Star")
plt.scatter(np.log(min_temp), np.log(max_mag), color="purple", marker="*", s=300, label="Dimmest Star")

plt.title("Hertzsprung-Russel Diagram", fontsize=16)
plt.xlabel("Temperature (log K)", fontsize=12)
plt.ylabel("Magnitude", fontsize=12)
plt.legend()

plt.show()
```
