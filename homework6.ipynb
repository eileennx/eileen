```python
matrix = np.array([
    [5, 12, 17, 9],
    [8, 21, 14, 3],
    [6, 13, 19, 11]
])
```

```python
shape = matrix.shape
print(f"Shape: {shape}")
print(f"Rows: {shape[0]}, Columns: {shape[1]}")
```

```python
print("Overall mean:", np.mean(matrix))
```

```python
print("Max in each row:", np.max(matrix, axis=1))
print("Min in each column:", np.min(matrix, axis=0))
```

```python
flat_matrix = matrix.flatten()
print("Flattened shape:", flat_matrix)
# The 2D array was collapsed into a single 1D array of 12 elements
```

```python
min_idx_flat = np.argmin(matrix)
max_idx_flat = np.argmax(matrix)
print(f"Flat matrix - Smallest Index: {min_idx_flat}, Largest Index: {max_idx_flat}")
```

```python
min_coords = np.unravel_index(min_idx_flat, matrix.shape)
max_coords = np.unravel_index(max_idx_flat, matrix.shape)
print(f"Flat matrix - Smallest Coordinates: {min_coords}, Largest Coordinates: {max_coords}")
```

```python
print("Values > 15:", matrix[matrix > 15])
```

```python
print("Sorted each row:\n", np.sort(matrix, axis=1))
```

```python
# --- Prime Star Systems ---
```

```python
arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])
```

```python
def is_prime(n):
    if n < 2: 
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def get_prime_rows(data):
    v_is_prime = np.vectorize(is_prime)
    prime_mask = v_is_prime(data)
    rows_with_primes = np.any(prime_mask, axis=1)
    return data[rows_with_primes]
```

```python
prime_star_systems = get_prime_rows(arr)
print("Star systems >= 1 prime magnitude:\n", prime_star_systems)
```

```python
# --- 5 Let’s Play Checkers! ---
```

```python
def checkerboard():
   # 5.1
    board = np.zeros((8,8), dtype = int)

    # 5.2
    board[0::2, 0::2] = 1

    # 5.3
    board[1::2, 1::2] = 1
    
    return board
```

```python
print("Standard Checkerboard:")
print(checkerboard())
```

```python
def reverse_checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[0::2, 1::2] = 1
    board[1::2, 0::2] = 1
    return board
```

```python
print("Reverse Checkerboard:")
print(reverse_checkerboard())
```

```python
# --- 6 The Expanding Universe ---
```

```python
def expansion(arr, n):
    def add_spaces(s):
        spacer = " " * n
        return spacer.join(s)
    
    v_expansion = np.vectorize(add_spaces)
    
    return v_expansion(arr)
```

```python
universe = np.array(['galaxy', 'clusters'])
print(expansion(universe, 1))
print(expansion(universe, 2))
```

```python
# --- 7 Second-Brightest Star ---
```

```python
def second_brightest(data):
    sorted_stars = np.sort(data, axis=1)
    return sorted_stars[:, -2]
```

```python
np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))
```

```python
result = second_brightest(stars)
print("Star Clusters:\n", stars)
print("\nSecond Brightest stars:", result)
```
